```python
#Feature Engineering-4 Assignment
"""Q1. What is data encoding? How is it useful in data science?"""
Ans: Data encoding is the process of transforming data from one format to another. It is useful in data 
science because it allows data to be stored and manipulated in a more efficient way. For example, encoding
categorical variables into numerical values can make it easier to perform calculations on the data. 
Additionally, encoding data into a more compact format can reduce the amount of storage space needed to 
store the data.

Data encoding can also be used to improve the accuracy of machine learning models. By encoding data into a 
format that is more suitable for the model, it can help the model better understand the data and make more
accurate predictions.

"""Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario."""
Ans: Nominal encoding is a type of data encoding that assigns a unique identifier to each category or 
value in a dataset. It is used to represent categorical variables, such as gender, race, or occupation. 
For example, in a survey about job satisfaction, you could use nominal encoding to assign a unique 
identifier to each job title. You could then use the identifiers to track responses from different job 
titles and analyze the data accordingly.

For example, you could assign the following identifiers to each job title:

1 - Manager 
2 - Engineer 
3 - Accountant 
4 - Analyst 
5 - Designer 
6 - Developer 
7 - Other

This would allow you to easily track responses from each job title and analyze the data accordingly.

"""Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example."""
Ans: Nominal encoding is preferred over one-hot encoding when the data is not ordinal and there is no natural 
ordering between the categories. For example, if you are collecting data on a person's favorite color, 
nominal encoding would be preferred over one-hot encoding since there is no natural ordering between the 
colors.

Nominal encoding would assign each color a numerical value, such as 1 for red, 2 for blue, 3 for green, 
etc. This allows the data to be stored and manipulated more easily than if one-hot encoding was used.

"""Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding
technique would you use to transform this data into a format suitable for machine learning algorithms?
Explain why you made this choice."""
Ans: I would use one-hot encoding(OHE) for this dataset. One-hot encoding is a technique used to transform 
categorical data into a numericalformat that can be used by machine learning algorithms. This technique 
creates a new column for each unique value in the dataset, with each column containing binary values 
(0 or 1) indicating whether the row contains that value. This allows the machine learning algorithm to
interpret the data as numerical values, rather than as categorical values.

One-hot encoding is a popular choice for transforming categorical data because it is simple to implement 
and does not require any assumptions about the relative importance of the different categories. 
Additionally, it does not introduce any bias into the data, which can be important when training a machine 
learning model.

"""Q5. In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns
are categorical, and the remaining three columns are numerical. If you were to use nominal encoding to
transform the categorical data, how many new columns would be created? Show your calculations."""
Ans: Nominal encoding creates a new column for each unique value in the categorical column. 

For the first categorical column, if there are 10 unique values, then 10 new columns will be created. 
For the second categorical column, if there are 15 unique values, then 15 new columns will be created. 

Therefore, the total number of new columns created by nominal encoding is 10 + 15 = 25.

"""Q6. You are working with a dataset containing information about different types of animals, including their
species, habitat, and diet. Which encoding technique would you use to transform the categorical data into
a format suitable for machine learning algorithms? Justify your answer."""

Ans: One-hot encoding would be the best technique to use for transforming the categorical data into a 
format suitable for machine learning algorithms. This is because one-hot encoding creates a binary vector 
for each category, which allows the algorithm to distinguish between different categories and assign 
weights accordingly. Additionally, one-hot encoding does not assume any order or hierarchy between the 
categories, which is important for machine learning algorithms.

This technique is also useful for datasets with a large number of categories, as it can reduce the 
dimensionality of the data and make it easier to work with. 

"""Q7.You are working on a project that involves predicting customer churn for a telecommunications
company. You have a dataset with 5 features, including the customer's gender, age, contract type,
monthly charges, and tenure. Which encoding technique(s) would you use to transform the categorical
data into numerical data? Provide a step-by-step explanation of how you would implement the encoding."""

Ans: 
Step 1: Identify the categorical features in the dataset. In this case, the categorical features are gender,
contract type, and age. 

Step 2: Decide which encoding technique to use for each feature. For gender, you could use binary encoding
(0 for male and 1 for female). For contract type, you could use one-hot encoding (creating a separate 
column for each contract type and assigning a 1 or 0 to indicate whether the customer has that contract 
type). For age, you could use ordinal encoding (assigning a numerical value to each age group).

Step 3: Implement the encoding. For binary encoding, you would create a new column for gender and assign 0
for male and 1 for female. For one-hot encoding, you would create a separate column for each contract type 
and assign a 1 or 0 to indicate whether the customer has that contract type. For ordinal encoding, you 
would create a new column for age and assign a numerical value to each age group (e.g. 0 for 18-25, 1 for
26-35, etc.).

Step 4: Validate the encoding. Once you have implemented the encoding, you should validate it by checking 
that the values in the new columns are correct and that the data is in the correct format.

Step 5: Save the encoded data. Once you have validated the encoding, you should save the encoded data in 
a new file or dataset.

Step 6: Use the encoded data for your analysis. Once you have saved the encoded data, you can use it for 
your analysis.

```
